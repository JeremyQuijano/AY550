The H and He fractions are essentially inverses of each other when looking at the center of a star. Stars convert Hydrogen to Helium, so as the star ages the amount of Hydrogen in the center will decrease and the amount of Helium in the center will increase. 

The difference in the plots for the 1 and 3 solar mass stars shows that the 3 solar mass star has used more of its Hydrogen, in the center and further from the center, than the 1 solar mass star has. This makes sense because higher mass stars will burn faster and begin using the Hydrogen further from the core before the 1 solar mass star. The 3 solar mass star also has a steeper slope than the 1 solar mass star. 

We would expect convection zones to have more homogenious distributions of matter. The constant "churning" of convection should create a mixture within the zone.

## (e) At what age does the $1 M_\odot$ star reach a central H fraction of $0.2$? How much energy is released for converting about $0.1 M_\odot$ from an H mass fraction of $0.75$ to $0.2$? How long does it take to radiate that much energy emitting a constant $1 L_\odot$? How far off is this simple esitmate of the age computed by MESA?

The star is roughly $6.2 \times 10^9$ years old when the central H fraction reaches 0.2.

It would take about $6.4 \times 10^9$ years to radiate that energy at $1 L_\odot$.

The difference in calcuation versus the MESA age is roughly $2.3 \times 10^8$ years

```python
# find the age of the star when H fraction = 0.2
age = np.max(history_1solarmass['star_age'])
print('1 solar mass star age:', format(age, '1.3E'), 'years')

# calculate mass being converted
year = 3.15E7 # seconds
mass_sun = 2.0E33 # grams
luminosity_sun = 3.8E33 # erg/s
energy_per_gram = 7.0E18 # erg/gram
center_H1_mass = 0.75 - 0.2 # mass fraction
total_mass_conversion = (0.1 * mass_sun) * center_H1_mass # grams
energy_released = energy_per_gram * total_mass_conversion # erg
time_nuclear = energy_released / luminosity_sun # seconds
time_nuclear_years = time_nuclear / year
print('Time nuclear:', format(time_nuclear_years, '1.3E'))
print('Difference:', format(time_nuclear_years - age, '1.3E'))
```

# MESA Plotting Functions

```python
# plot one profile on one plot
def plot_single_profile(x, y, profile, label, title, xlabel, ylabel, xscale, yscale):
    """
    PLOT ONE PROFILE
    
    x       = x-axis data
    y       = y-axis data
    profile = index number of profile to be plotted
    label   = name of array with sim data
    title   = title for plot
    xlabel  = x-axis label
    ylabel  = y-axis label
    xscale  = x-axis scale
    yscale  = y-axis scale
    
    Will return a plot for a specified profile
    """
    
    plt.plot(label[profile][1][x], label[profile][1][y], label=label[profile][0])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xscale(xscale)
    plt.yscale(yscale)
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    return(plt.show())

# plot all profiles on one plot
def plot_group_together(x, y, label, title, xlabel, ylabel, xscale, yscale):
    """
    PLOT ALL PROFILES ON ONE PLOT
    
    x      = x-axis data
    y      = y-axis data
    label  = name of array with sim data
    title  = title for plot
    xlabel = x-axis label
    ylabel = y-axis label
    xscale = x-axis scale
    yscale = y-axis scale
    
    Will return a single plot will all profiles plotted
    """
    
    for i in range(len(label)):
        plt.plot(label[i][1][x], label[i][1][y], label=label[i][0])
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.xscale(xscale)
        plt.yscale(yscale)
        plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    return(plt.show())

# plot two profiles on one plot
def plot_two_together(x, y, label1, label2, profile_1, profile_2, title, xlabel, ylabel, xscale, yscale):
    """
    PLOT ALL PROFILES ON ONE PLOT
    
    x        = x-axis data
    y        = y-axis data
    label1    = name of array with sim data 1
    label2    = name of array with sim data 2
    profile_1 = profile for sim data 1
    profile_2 = profile for sim data 2
    title     = title for plot
    xlabel    = x-axis label
    ylabel    = y-axis label
    xscale    = x-axis scale
    yscale    = y-axis scale
    
    Will return a single plot will all profiles plotted
    """
    
    plt.plot(label1[profile_1][1][x], label1[profile_1][1][y], label=label1[profile_1][0])
    plt.plot(label2[profile_2][1][x], label2[profile_2][1][y], label=label2[profile_2][0])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xscale(xscale)
    plt.yscale(yscale)
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    return(plt.show())


# plot all profiles on individual plots
def plot_group_individual(x, y, label, title, xlabel, ylabel, xscale, yscale):
    """
    PLOT ALL PROFILES ON INDIVIDUAL PLOTS
    
    x      = x-axis data
    y      = y-axis data
    label  = name of array with sim data
    title  = title for plot
    xlabel = x-axis label
    ylabel = y-axis label
    xscale = x-axis scale
    yscale = y-axis scale
    
    Will return individual plots for each profile
    """
    
    for i in range(len(label)):
        plt.plot(label[i][1][x], label[i][1][y], label=label[i][0])
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.xscale(xscale)
        plt.yscale(yscale)
        plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
        plt.show()
    return(plt.show())

# plot history 
def plot_history(x, y, label, title, xlabel, ylabel, xscale, yscale):
    """
    PLOT HISTORY
    
    x      = x-axis data
    y      = y-axis data
    label  = name of array with sim data
    title  = title for plot
    xlabel = x-axis label
    ylabel = y-axis label
    xscale = x-axis scale
    yscale = y-axis scale
    
    Will return plot with desired history data
    """
    
    plt.plot(label[x], label[y])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xscale(xscale)
    plt.yscale(yscale)
    return(plt.show())
```

# MESA Data Loader Function

```python
def data_loader(path):
    """
    DATA LOADER USING GENFROMTXT
    
    How to call:
    path = 'some/path/to/directory'
    array_name = data_loader(path)
    
    Parameters:
    path = string path to file
    
    Function:
    Will return an array with desired data
    """
    
    return np.genfromtxt(path, names=True, skip_header=5)
```

# MESA Profile Loader Function

```python
def profile_loader(path, array_name, sim_name):
    """
    PROFILE DATA LOADER USING GENFROMTXT
    
    How to call:
    path = 'some/path/to/directory'
    array_name = []
    profile_loader(path, array_name, sim_name)
    
    Parameters:
    path       = string path to desired directory
    array_name = name of initialized array
    sim_name   = name of sim/experiment
    
    Details:
    glob.glob     = easily find desired files
    sorted(, key) = order files ascending
    replace       = create name for each profile
    
    Array indexing [][][]:
    array_name[i][0]             = sim name profile#
    array_name[i][1].dtype       = array with profile headers
    array_name[i][1]['quantity'] = array of desired quanitity from specified profile
    
    Function:
    Will return return an array with MESA profiles
    """
    
    for file in sorted(glob.glob(path + 'profile*.data'), key=os.path.getmtime):
        array_name.append([file.replace(path, sim_name).replace('.data', ''), 
                           data_loader(file)])
    return array_name
```
