# (c) How does the $T_\text{eff}$ change in the giant phases and in the WD phase compared to the core hydrogen burning phase? How is this related to the star's radius?

The Hydrogen burning phase has a constant $T_\text{eff}$. The star will then slightly decrease in $T_\text{eff}$ for the Red Giant Phase, until it begins to heat up again for the Asymptotic Giant Phase. The $T_\text{eff}$ will again decrease until the star begins to die. It then blows off the outer atmosphere and reveals the core of the remnant White Dwarf, which has a much higher $T_\text{eff}$ than the previous phases.

Phase | $T_\text{eff}$ | Radius
:-----:|:-----:|:-----:
Hydrogen Burning | Constant | Constant
Red Giant | Decreases slightly | Increases
Asymptotic Giant | Increases slightly | Decreases
White Dwarf | Increases dramatically | Decreases dramatically

## Helper Functions

```python
def data_loader(path):
    """
    DATA LOADER USING GENFROMTXT
    
    How to call:
    path = 'some/path/to/directory'
    array_name = data_loader(path)
    
    Parameters:
    path = string path to file
    
    Function:
    Will return an array with desired data
    """
    
    return np.genfromtxt(path, names=True, skip_header=5)
```

```python
def profile_loader(path, array_name, sim_name):
    """
    PROFILE DATA LOADER USING GENFROMTXT
    
    How to call:
    path = 'some/path/to/directory'
    array_name = []
    profile_loader(path, array_name, sim_name)
    
    Parameters:
    path       = string path to desired directory
    array_name = name of initialized array
    sim_name   = name of sim/experiment
    
    Details:
    glob.glob     = easily find desired files
    sorted(, key) = order files ascending
    replace       = create name for each profile
    
    Array indexing [][][]:
    array_name[i][0]             = sim name profile#
    array_name[i][1].dtype       = array with profile headers
    array_name[i][1]['quantity'] = array of desired quanitity from specified profile
    
    Function:
    Will return an array with MESA profiles
    """
    
    for file in sorted(glob.glob(path + 'profile*.data'), key=os.path.getmtime):
        array_name.append([file.replace(path, sim_name).replace('.data', ''), 
                           data_loader(file)])
    return array_name
```
